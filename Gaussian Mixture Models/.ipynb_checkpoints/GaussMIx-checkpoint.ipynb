```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import multivariate_normal
import scipy.stats
```

```python
#Function for generating 1 dimensional data
#Accepts n, the number of samples from each gaussian
def gen1D(n):
    #mus = np.random.uniform(1.5,8.5,3)
    stdevs = np.random.uniform(.3,1,3)
    
    g0 = np.random.normal(1.72,stdevs[0],n)
    g1 = np.random.normal(5.26,stdevs[1],n)
    g2 = np.random.normal(7.91,stdevs[2],n)

    #g0 = np.random.normal(2.3,.5,n)
    #g1 = np.random.normal(8.7,.7,n)
    #g2 = np.random.normal(6.1,1.2,n)
    
    g = np.append(g0,g1)
    g = np.append(g,g2)
    g = np.random.permutation(np.ravel(g))
    return g
```

```python
#Applies the EM Algorithm on a data vector d, with assumed number of classes k,
#number of permutations n, and guessed mu, cov, and pi vectors.
def EMAlg(d,k,n,mus,sigs,pis):
    n
    nd = np.size(d)
    resp = np.zeros((nd,k))
    sigs = np.array(sigs)
    mus = np.array(mus)
    pis = np.array(pis)
    sigs = sigs*sigs
    
    for i in range(n):
        
        for j in range(nd):
            p = 0
            for m in range(k):
                p = p + (pis[m]*scipy.stats.norm(mus[m],sigs[m]).pdf(d[j]))
            for n in range(k):
                resp[j,n] = (pis[n]*scipy.stats.norm(mus[n],sigs[n]).pdf(d[j]))/p
                
        for l in range(k):
            mus[l] = (1/np.sum(resp[:,l])) * np.sum(np.atleast_2d(resp[:,l])*np.atleast_2d(d))
            mulong = np.ones((1,nd))*mus[l]
            sigs[l] = ((1/np.sum(resp[:,l])) * np.sum(np.atleast_2d(resp[:,l]).T@
                                                     (np.atleast_2d(d)-mulong)@
                                                     (np.atleast_2d(d)-mulong).T))
            pis[l] = (np.sum(resp[:,l]))/nd
        
    return mus, np.sqrt(sigs), pis
```

```python
#Plots histogram of the data points
def plotHist(a,n,ax):
    h = np.histogram(a,bins=80,range=(0,10))
    h0 = np.array(h[0])/(n/3)
    h1 = np.array(h[1])
    ax.bar(h1[:80],h0,.25,align='edge',color='grey')
```

```python
#Plots pdfs of three gaussians
def plotPDFs(mu,sigma,ax):
    x = np.linspace(0,10,100)
    
    ax.plot(x,mlab.normpdf(x, mu[0], sigma[0]))
    ax.plot(x,mlab.normpdf(x, mu[1], sigma[1]))
    ax.plot(x,mlab.normpdf(x, mu[2], sigma[2]))
```

```python
#Generation and plotting
n = 100
g = gen1D(n)
figure, ax = plt.subplots(nrows=2, ncols=2, figsize=[16,8])

k = 3
muGuess = [1.5,5,7.5]
sigGuess = [1,1,1]
piGuess = [.33,.33,.33]

plotPDFs(muGuess,sigGuess,ax[0,0])

mu1,sig1,pi1 = EMAlg(g,k,1,muGuess,sigGuess,piGuess)
plotPDFs(mu1,sig1,ax[0,1])
print(mu1)
print(sig1)
print(pi1)

plotHist(g,n,ax[0,0])
plotHist(g,n,ax[0,1])
plotHist(g,n,ax[1,0])
plotHist(g,n,ax[1,1])

plt.show()
```
